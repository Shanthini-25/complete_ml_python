##Visualization

```python
sns.pairplot(df,hue="Class",height=3,aspect=1)
```

```python
sns.pairplot(iris,hue="species",height=3,aspect=1,diag_kind="hist")
#data, hue-->Variable in data to map plot aspects to different colors, height-->Height (in inches) of each facet, aspect-->Aspect * height gives the width (in inches) of each facet
```

```python
df.hist(edgecolor="black",linewidth=1,figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
#to get a grid like appearance
plt.subplot(2,2,1)
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='sepal_width',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='petal_length',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='petal_width',data=iris)
```

```python
sns.violinplot(x='species',y='sepal_width',data=iris)
```
